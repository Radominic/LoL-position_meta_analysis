## 격차를 더해 격차 총량 구하기

```python
import pandas as pd
import numpy as np
dataGoldD = pd.DataFrame(columns = ['gameId','TOP','MIDDLE','BOTTOM','JUNGLE'])
dataXpD = pd.DataFrame(columns = ['gameId','TOP','MIDDLE','BOTTOM','JUNGLE'])
index = 0
for game in dataGold.keys():
    tempG = [0,0,0,0]
    tempX = [0,0,0,0]
    for row in range(0,20):
        tempG+=  dataGold[game].iloc[[row]].values[0]
        tempX+=  dataXp[game].iloc[[row]].values[0]
    tempG = np.insert(tempG,0,game)
    tempX = np.insert(tempX,0,game)
    
    dataGoldD.loc[index] = tempG
    dataXpD.loc[index] = tempX
    index += 1
```
